Sort the coins based on their volumes

There are 2 ways to do that either by making a new column indicating the rank of each coin, or sorting the positions of the coins within their column (we must use the coins as the second index using the .stack())

We are going to use the sorting method, unless we need the numerical value of the rank

```python
#By Rank
# Reset index to flatten the MultiIndex
# df_reset = data.reset_index()

# Rank the 'volume' column within each group (group by first and second index levels)
data['rank'] = data['volume'].groupby(data.index.get_level_values(0)).rank(ascending=False)

# # Set the index back to the original MultiIndex (optional)
# df_ranked = df_reset.set_index([df_reset.columns[0], df_reset.columns[1]])
# df_ranked
data
```

```python
#By sorting
# Reset index to flatten the MultiIndex
df_reset = data.reset_index()

# Sort the data by 'volume' within each date group
df_sorted = df_reset.sort_values(by=[df_reset.columns[0], 'volume'], ascending=[True, False])

# Set the index back to the original MultiIndex (optional)
df_sorted = df_sorted.set_index([df_sorted.columns[0], df_sorted.columns[1]])
df = df_sorted
df.index.names = ['date', 'coin']
data = df
```

```python
#Making a function to sort based on the volume
def sort_by_volume(df):
    df_reset = df.reset_index()
    df_sorted = df_reset.sort_values(by=[df_reset.columns[0], 'volume'], ascending=[True, False])
    df_sorted = df_sorted.set_index([df_sorted.columns[0], df_sorted.columns[1]])
    return df_sorted
```

```python
#Make it as a function
def universe(df):
    universe = []
    for coin in df.unstack().columns.levels[1]:
        if coin not in universe:
            universe.append(coin)

    return universe
```
