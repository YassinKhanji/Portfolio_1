```python
#Now Make it as a function
def above_ema(data, df, ema_window):
    """
    This assumes a multi-index dataframe with 'close' as the column level 0 and the symbols as the column level 1
    """
    df, data = df.copy().unstack(), data.copy().unstack()
    for coin in data.columns.get_level_values(1).unique():
        df[f'ema_{ema_window}', coin] = ta.ema(data['close', coin], length=ema_window)
    df = df.stack(future_stack = True)
    df['above_ema'] = (df['close'] > df[f'ema_{ema_window}']).astype(int)
    return df
```

```python
df = above_ema(data, 2)
```
