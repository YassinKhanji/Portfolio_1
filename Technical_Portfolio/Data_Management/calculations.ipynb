```python
def trades(df):
    df = df.unstack()

    for coin in df['close'].columns:
        df['trades', coin] = df['position', coin].diff().fillna(0).abs()

    #A case where we start with a position, we need to add a trade
    for coin in df['close'].columns:
        if df['position', coin].iloc[0] ==1:
            first_date = df.index[0]
            df.loc[first_date, ('close', coin)]= 1
    df = df.stack(future_stack=True)
    return df
```

```python
for coin in df['close'].columns:
    df['position', coin] = np.where(df['close', coin].shift(1) > df['close', coin], 1, 0)
    df["strategy", coin] = df["position", coin].shift(1) * df["returns", coin]
    df['trades', coin] = df['position', coin].diff().fillna(0).abs()
```

```python
df = df.sort_index(axis=1, level=0) 
df
```

```python
for coin in df.columns.levels[1]:
    df['session', coin] = np.sign(df['trades', coin]).cumsum().shift().fillna(0)
df.head(10)
```

Putting everything together

```python
def sessions(df):
    _df = df.copy().unstack()
    for coin in _df['close'].columns:
        _df['session', coin] = np.sign(_df['trades', coin]).cumsum().shift().fillna(0)
        _df[('session_return', 'BTCUSDT')] = _df['strategy', 'BTCUSDT'].groupby(_df['session', coin]).cumsum().apply(np.exp) - 1
        _df[('session_compound', coin)] = _df['strategy', coin].groupby(_df['session', coin]).cumsum().apply(np.exp) - 1
    
    _df = _df.stack(future_stack=True)

    return _df
```

```python
import pandas as pd

# Example DataFrame
data = {
    'sessions': [1, 1, 1, 2, 2, 3, 3, 3],
    'close': [100, 105, 110, 120, 115, 130, 135, 140]
}
df = pd.DataFrame(data)

# Calculate returns
df['returns'] = df['close'].pct_change()

# Calculate session returns
df['session_return'] = df.groupby('sessions')['close'].pct_change().fillna(0) #This is to get the return row by row
df['overall_session_return'] = df.groupby('sessions')['returns'].transform(lambda x: (x + 1).prod() - 1) #This is to get the return of the whole session
df['first_close'] = df.groupby('sessions')['close'].transform(lambda x: x.iloc[0]) #Or we can type first
df
```
